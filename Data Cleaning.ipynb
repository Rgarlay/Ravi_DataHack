```python
df1.columns
list1 = []
for i in df1.columns:
    list1.append(i)
```

```python
list2 = list1[:20]
list3 = list1[20]
```

```python
list2
for i in list2:
    abc = df1[i].mode().iloc[0]
    df1.fillna({i:abc}, inplace=True)
```

```python
df1['education'] = df1['education'].map({'< 12 Years': 1, '12 Years': 2 ,'Some College': 3,'College Graduate': 4})
```

```python
df1.fillna({'education' : 2}, inplace=True)
```

```python
df1 = pd.get_dummies(df1 , prefix = 'edu', columns = ['education'])
```

```python
df1['age_group'] = df1['age_group'].map({'18 - 34 Years' : 0, '35 - 44 Years': 1,'45 - 54 Years' : 1,'55 - 64 Years': 1,'65+ Years': 2})
```

```python
df1['sex'] = df1['sex'].map({'Male':0, 'Female':1})
```

```python
df1 = pd.get_dummies(df1 , prefix = 'ag', columns = ['age_group'])
```

```python
df1['income_poverty'] = df1['income_poverty'].map({'Below Poverty': 1 ,'<= $75,000, Above Poverty': 2 ,'> $75,000': 3}) 
```

```python
df1.fillna({'income_poverty':2}, inplace=True)
```

```python
rent_condition = df1['income_poverty'].isin([1, 2])
owned_condition = df1['income_poverty'] == 3

# Correct the .isnull() method and fill 'rent_or_own' column based on conditions
df1.loc[df1['rent_or_own'].isnull() & rent_condition, 'rent_or_own'] = 'Own'
df1.loc[df1['rent_or_own'].isnull() & owned_condition, 'rent_or_own'] = 'Rent'
```

```python
df1 = pd.get_dummies(df1 , prefix = 'inc', columns = ['income_poverty'])
```

```python
df1['marital_status'] = df1['marital_status'].map({'Married':2,'Not Married':1})
```

```python
df1.fillna({'marital_status':1}, inplace=True)
```

```python
df1 = pd.get_dummies(df1 , prefix = 'wed', columns = ['marital_status'])
```

```python
df1['rent_or_own'] = df1['rent_or_own'].map({'Rent': 0,'Own':1}) 
```

```python
df1['employment_status'] = df1['employment_status'].map({'Employed': 1,'Not in Labor Force':2,'Not Employed': 3}) 
```

```python
df1.fillna({'employment_status': 2}, inplace=True)
```

```python
df1 = pd.get_dummies(df1 , prefix = 'emp', columns = ['employment_status'])
```

```python
df1 = pd.get_dummies(df1, columns=['race'], prefix='d')
```

```python
df1.fillna({'household_children': df1['household_children'].mode()[0]}, inplace=True)
df1.fillna({'household_adults': df1['household_adults'].mode()[0]}, inplace=True)
```

```python
df1 = pd.get_dummies(df1, columns=['hhs_geo_region'], prefix='town')
```

```python
df1 = pd.get_dummies(df1, columns=['census_msa'], prefix='region')
```
